```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# load pima indians dataset
dataset = numpy.loadtxt("pima-indians-diabetes.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=8, init='uniform', activation='relu')) 
model.add(Dense(8, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) # Fit the model
model.fit(X, Y, validation_split=0.33, nb_epoch=150, batch_size=10)
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=seed)
```

```python
# create model
model2 = Sequential()
model2.add(Dense(12, input_dim=8, init='uniform', activation='relu'))
model2.add(Dense(8, init='uniform', activation='relu'))
model2.add(Dense(1, init='uniform', activation='sigmoid'))
# Compile model
model2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model
model2.fit(X_train, y_train, validation_data=(X_test,y_test), nb_epoch=150, batch_size=10)
```

```python
from sklearn.cross_validation import StratifiedKFold
```

```python
# define 10-fold cross validation test harness
kfold = StratifiedKFold(y=Y, n_folds=10, shuffle=True, random_state=seed)
cvscores = []
for i, (train, test) in enumerate(kfold):
# create model
    model3 = Sequential()
    model3.add(Dense(12, input_dim=8, init='uniform', activation='relu')) 
    model3.add(Dense(8, init='uniform', activation='relu'))
    model3.add(Dense(1, init='uniform', activation='sigmoid'))
# Compile model
    model3.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) # Fit the model
    model3.fit(X[train], Y[train], nb_epoch=150, batch_size=10, verbose=0)
# evaluate the model
    scores = model3.evaluate(X[test], Y[test], verbose=0)
    print("%s: %.2f%%" % (model3.metrics_names[1], scores[1]*100)) 
    cvscores.append(scores[1] * 100)
print("%.2f%% (+/- %.2f%%)" % (numpy.mean(cvscores), numpy.std(cvscores)))
```
